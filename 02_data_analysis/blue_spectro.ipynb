# Custom Colors Map

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
#import matplotlib.colors
#from matplotlib.colors import CSS4_COLORS
import numpy as np
import librosa
import librosa.display
colorlist=["midnightblue", "mediumblue", "dodgerblue", "azure"]
newcmp = LinearSegmentedColormap.from_list('testCmap', colors=colorlist, N=256)
```

```python
signal , sr = librosa.load('/home/bxnxne/Téléchargements/BE3C_7501000C.wav')
S = librosa.feature.melspectrogram(y=signal, sr=44100) # Compute a mel-scaled spectrogram
S_DB = librosa.power_to_db(S, ref=np.max)

librosa.display.specshow(S_DB, x_axis='time',y_axis='mel', sr=44100,fmax=8000,cmap=newcmp);
plt.savefig('foo.png')
plt.show()
```
