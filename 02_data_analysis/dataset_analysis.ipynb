```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load a dataset

```python
path_to_dataset = '/content/drive/MyDrive/lewagon-deepdive/raw_data/watkins_sound_list_extended.csv'
dataset = pd.read_csv(path_to_dataset)
dataset = dataset.rename(columns={"species_name_x": "species_name"})
dataset
```

```python
dataset.shape
```

# Check balance of categories

## Count of categories

```python
plt.figure(figsize=(10,10))
sns.barplot(x=dataset['species_code'].value_counts(), y=dataset['species_code'].unique());
```

## Duration of categories

```python
df_duration = dataset.groupby(['species_code']).sum()
df_duration.sort_values(by='duration', ascending=False, inplace = True)


plt.figure(figsize=(10,10))
sns.barplot(x=df_duration.duration, y=df_duration.index);
```
