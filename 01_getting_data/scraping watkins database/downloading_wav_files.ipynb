## Concatenating all csv files in one df

```python
best_species = [
'AA1A','AA3A','AA3B','AC1A','AC1F','AC2A','BA2A','BB1A','BB2A','BD3B','BD4A','BD5A','BD6A','BD6B',
'BD10A','BD15A','BD15B','BD15C','BD15F','BD15L','BD17A','BD19D','BE3C','BE3D','BE7A','BE9A','CB1A',
'CC2A','CC4A','CC5A','CC12G','CC14A']
```

```python
!pwd
```

```python
df_list = []
for species in best_species:
    df_list.append(pd.read_csv(project_path + f'data/csv_files/best_of_{species}.csv', index_col=0, on_bad_lines='warn',sep=';'))
```

```python
best_df = pd.concat(df_list,ignore_index=True)
best_df.shape
```

```python
best_df.to_csv(project_path + 'data/best_df.csv',sep=';')
```

## Downloading all wav files

```python
best_df.iloc[0]['species_code'] + "_" + best_df.iloc[0]['download_link'].split('/')[-1]
```

```python
root_url = 'https://whoicf2.whoi.edu'
```

```python
for index, row in tqdm(best_df.iloc[1500:].iterrows()):
    
    time.sleep(random.uniform(2, 4)) #time sleep in order not to overload server
    
    tail_url = row['download_link']
    
    file_name = row['species_code'] + "_" + row['download_link'].split('/')[-1]
    
    doc = requests.get(root_url+tail_url)
    
    with open(f'best_of_cuts/wav_files/{file_name}','wb') as f:
        f.write(doc.content)

print("scraping done")
```
